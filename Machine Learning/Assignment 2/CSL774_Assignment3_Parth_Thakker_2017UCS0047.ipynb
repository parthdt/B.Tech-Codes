## Tasks for building a Recommendation Model

**Some of the movies hadnâ€™t been watched and therefore, are not rated by the users. Netflix would like to take this as an opportunity and build a machine learning recommendation algorithm which provides the ratings for each of the users.**

### Task 4 - Divide the data into training and test data

**First, I fill null values with 0, invert the dataframe to get a `user_id` vs `Movie Number` dataframe, and then divide the data.**

```
data = data.fillna(0)
data
```

```
data_inverted = pd.melt(data,id_vars='user_id')
data_inverted.rename(columns={'variable':'Movie Number', 'value':'Rating'}, inplace=True)
data_inverted
```

```
data_inverted[data_inverted['Movie Number'] == 'Movie69']   #Random movie
```

```
from surprise import Reader, Dataset
from surprise.model_selection import train_test_split

reader = Reader()
model_data = Dataset.load_from_df(data_inverted[['user_id','Movie Number','Rating']], reader)

train, test = train_test_split(model_data, test_size = 0.2)
```

### Task 5 - Build a recommendation model on training data

**Using a SVD model for this task.**

```
from surprise import SVD, accuracy
model = SVD()
model.fit(train)
```

### Task 6 - Make predictions on the test data

**For this task, I use the `test()` method for prediction on test data, and show the root mean square error using `rmse()` method of accuracy module.**

```
predicted_values = model.test(test)   #Get list of predictions
predicted_values_df = pd.DataFrame(predicted_values)
predicted_values_df.rename(columns={'iid':'Predicted Movie Number', 'uid':'user_id'}, inplace=True)
predicted_values_df
```

```
accuracy.rmse(predicted_values)   #Calculate root mean square error
```
