### Imports


```
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Amazon - Movies and TV Ratings.csv')
data.head()
```

### Task 1 - Which movies have maximum views/ratings?

```
movie_notNAN = pd.DataFrame()
movie_notNAN['Number of Ratings'] = data.notnull().sum(axis=0).drop('user_id')
print("Movie with Maximum Views/Ratings:\n")
movie_notNAN[movie_notNAN['Number of Ratings'] == movie_notNAN['Number of Ratings'].max()]
```
